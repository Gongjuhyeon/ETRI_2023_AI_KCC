```python
import neurokit2 as nk
```

```python
ecg_dir = "../../dataset/KEMDy19/ECG"
```

```python
ecg_dir
```

```python
ecg_csv_files = []
for session_dir in glob.glob(os.path.join(ecg_dir, "Session*", "Original")):
  session_files = glob.glob(os.path.join(session_dir, "Sess*.csv"))
  ecg_csv_files.extend(session_files)
```

```python
for ecg_file in ecg_csv_files:
  # 현재 파일 이름 추출
  current_ecg_file_name = ecg_file.split('/')[-1]

  # Converting csv to dataframe
  ecg_csv_lines = []
  with open(ecg_file, 'r') as file:
    for line in file:
      values = line.strip().split(',')
      ecg_csv_lines.append(values)
  ecg_temp_df = pd.DataFrame(ecg_csv_lines)
  ecg_temp_df = ecg_temp_df.iloc[1:, :]
  ecg_temp_df.columns = ['Time_cum', 'ECG', 'Time', 'Speaker']

  # 형변환
  ecg_temp_df['ECG'] = ecg_temp_df['ECG'].astype(float)
  ecg_temp_df['Time'] = pd.to_datetime(ecg_df['Time'], format='%H:%M:%S.%f')
  ecg_temp_df['Time_cum'] = ecg_temp_df['Time_cum'].astype(float)
  ecg_temp_df.reset_index(drop=True, inplace=True)

  # Converting ECG to IBI
  neurokit2_temp = nk.ecg_process(ecg_temp_df['ECG'], 250)
  r_peaks_temp = neurokit2_temp[1]['ECG_R_Peaks']
  heart_beat_temp = ecg_temp_df.iloc[r_peaks_temp, :]
  ecg2ibi_temp_df = heart_beat_temp.reset_index(drop=True)
  ecg2ibi_temp_df['IBI'] = 0
  # ECG의 r-peak 사이 간격을 토대로 IBI 추출
  for idx, row in ecg2ibi_temp_df.iterrows():
    if idx != 0:
      ecg2ibi_temp_df['IBI'][idx] = (ecg2ibi_temp_df.iloc[idx, 2] - ecg2ibi_temp_df.iloc[idx-1, 2]).total_seconds()
  # Column & Row rearrangement
  ecg2ibi_temp_df = ecg2ibi_temp_df[['Time_cum', 'IBI', 'Time', 'Speaker', 'ECG']]
  ecg2ibi_temp_df = ecg2ibi_temp_df[['Time_cum', 'IBI', 'Time', 'Speaker']]
  ecg2ibi_temp_df = ecg2ibi_temp_df.iloc[1:, :]

  #현재 파일을 IBI_NEW 폴더에 .csv 형식으로 저장
  ecg2ibi_temp_df.to_csv(f'.../../dataset/KEMDy19/IBI_NEW/{current_ecg_file_name}')

  print(f'{current_ecg_file_name} ... Done!')
```

fleiss kappa - 함수 생성


```python
import numpy as np
import random

def fleiss_kappa(ratings, n_raters):
    n_items, n_categories = ratings.shape
    
    p = ratings.sum(axis=0) / (n_items * n_raters)
    P = ((ratings * ratings).sum(axis=1) - n_raters) / (n_raters * (n_raters - 1))
    P_mean = P.mean()
    P_expected = (p * p).sum()
    # print(f'P_mean : {P_mean:.4f}\nP_expected : {P_expected:.4f}')
    kappa = (P_mean - P_expected) / (1 - P_expected)
    
    return kappa


def p_i_score(ratings, n_raters):

    return ((ratings * ratings).sum(axis=1) - n_raters) / (n_raters * (n_raters - 1))


def build_ratings_matrix(evaluator_ratings, n_items, n_categories):
    ratings = np.zeros((n_items, n_categories))
    
    for item_index, category in evaluator_ratings:
        ratings[item_index][category-1] += 1
        
    return ratings


def get_p_i(evaluator_ratings, n_categories, n_raters):
    ratings_matrix = build_ratings_matrix(evaluator_ratings, 1, n_categories)
    p_i = p_i_score(ratings_matrix, n_raters)

    return float(p_i)


def get_kappa_score(evaluator_ratings, n_items, n_categories, n_raters):
    ratings_matrix = build_ratings_matrix(evaluator_ratings, n_items, n_categories)
    kappa = fleiss_kappa(ratings_matrix, n_raters)
    # print(f"Fleiss' Kappa: {kappa:.4f}")
    return kappa
```

```python
# 그룹 내 편차를 계산하기 위한 함수
def get_difference(arr):
    if len(arr) <= 1:
        return np.nan
    else:
        return arr.iloc[-1] - arr.iloc[0]
```

```python
def signal_processing(signal_df, signal):

  signal_df[signal] = signal_df[signal].astype(float) # 형변환
  signal_df['Time'] = pd.to_datetime(signal_df['Time'])
  try:
    signal_df['Time_cum'] = signal_df['Time_cum'].astype(float)
  except:
    pass
  grouped_by_speaker = signal_df.groupby('Speaker')[signal].agg(['count', 'mean', 'std', 'max', 'min', get_difference])
  grouped = pd.DataFrame({'Speaker': grouped_by_speaker.index, f'{signal}_mean': grouped_by_speaker['mean'], \
                           f'{signal}_count': grouped_by_speaker['count'], f'{signal}_max': grouped_by_speaker['max'], f'{signal}_min':grouped_by_speaker['min'],\
                           f'{signal}_difference': grouped_by_speaker[get_difference.__name__], f'{signal}_std': grouped_by_speaker['std']})
  grouped[f'{signal}_range'] = grouped[f'{signal}_max'] - grouped[f'{signal}_min']
  grouped.reset_index(drop=True, inplace=True)
  grouped = grouped.reindex(columns=['Speaker', f'{signal}_mean', f'{signal}_count', f'{signal}_max', f'{signal}_min', f'{signal}_std', f'{signal}_difference', f'{signal}_range'])
  
  return grouped
```

```python
for i in range(40):
  # Load IBI files
  ibi_19_ = pd.read_csv(ibi_19_csv_files[i])
  ibi_19_grouped_ = signal_processing(ibi_19_, 'IBI')

  # Load EDA files
  eda_19_csv_lines = []
  with open(eda_19_csv_files[i], 'r') as file:
    for line in file:
      values = line.strip().split(',')
      eda_19_csv_lines.append(values)
  eda_19 = pd.DataFrame(eda_19_csv_lines)
  eda_19 = eda_19.iloc[1:, :]
  eda_19.columns = ['EDA', 'Time_cum', 'Time', 'Speaker']
  eda_19_grouped = signal_processing(eda_19, 'EDA')
  
  # Load TEMP files
  temp_19_csv_lines = []
  with open(temp_19_csv_files[i], 'r') as file:
    for line in file:
      values = line.strip().split(',')
      temp_19_csv_lines.append(values)
  temp_19 = pd.DataFrame(temp_19_csv_lines)
  temp_19 = temp_19.iloc[1:, :]
  temp_19.columns = ['TEMP', 'Time_cum', 'Time', 'Speaker']
  temp_grouped = signal_processing(temp_19, 'TEMP')

  # Load Annotation files
  annotation_19 = pd.read_csv(annotation19_csv_files[i])
  annotation_19_refined = annotation_19.iloc[1:, 1:]
  annotation_19_refined.rename(columns=(new_columns_00 if i < 20 else new_columns_10), inplace=True)
  annotation_19_refined[['Wav_start', 'Wav_end']] = annotation_19_refined[['Wav_start', 'Wav_end']].astype(float)
  if i < 20:
    for e in evaluator_00:
      annotation_19_refined[f'{e}_emotion'] = annotation_19_refined[f'{e}_emotion'].map(emotion_category_map)
  else:
    for e in evaluator_10:
      annotation_19_refined[f'{e}_emotion'] = annotation_19_refined[f'{e}_emotion'].map(emotion_category_map)
  for idx, row in annotation_19_refined.iterrows():
    # Get P_i value of emotion evaluations
    emo_score = list(annotation_19_refined.loc[idx, emotion_col_00 if i < 20 else emotion_col_10])
    emo_tuple = [(0, e) for e in emo_score]
    annotation_19_refined.loc[idx, 'Emotion_p_i'] = get_p_i(emo_tuple, 7, 10)

    # Get Fleiss-Kappa score of Valence & Arousal
    valence_score = list(map(int, list(annotation_19_refined.loc[idx, valence_col_00 if i < 20 else valence_col_10])))
    arousal_score = list(map(int, list(annotation_19_refined.loc[idx, arousal_col_00 if i < 20 else arousal_col_10])))
    valence_tuple = [(0, e) for e in valence_score]
    arousal_tuple = [(1, e) for e in arousal_score]
    annotation_19_refined.loc[idx, 'Fleiss_Kappa'] = get_kappa_score(valence_tuple + arousal_tuple, 2, 5, 10)
  annotation_19_refined = annotation_19_refined[['Wav_start', 'Wav_end', 'Speaker', 'Emotion', 'Valence', 'Arousal', 'Emotion_p_i', 'Fleiss_Kappa']]

  
  # Merge files
  merging_ibi_df = pd.merge(annotation_19_refined, ibi_19_grouped_, on='Speaker', how='left')
  merging_eda_df = pd.merge(merging_ibi_df, eda_19_grouped, on='Speaker', how='left')
  merged_df = pd.merge(merging_eda_df, temp_grouped, on='Speaker', how='left')
  
  # Add Session column
  current_file_name = annotation19_csv_files[i].split('/')[-1]
  current_file_name = current_file_name.split('.')[0]
  merged_df['Session'] = current_file_name

  # Concat to final dataframe
  final_19_df = pd.concat([final_19_df, merged_df], ignore_index=True)

  print(f'{current_file_name} ... Done!')
```

```python
ibi_20_dir = "../../dataset/KEMDy20/IBI"
eda_20_dir = "../../dataset/KEMDy20/EDA"
temp_20_dir = "../../dataset/KEMDy20/TEMP"
annotation_20_dir = "../../dataset/KEMDy20/annotation"

ibi_20_csv_array = []
eda_20_csv_array = []
temp_20_csv_array = []
annotation_20_files = glob.glob(os.path.join(annotation_20_dir, "Sess*.csv"))
annotation_20_files = sorted(annotation_20_files)


# 파일 불러오기
for i in range(40):
  #ibi
  ibi_20_csv_files = glob.glob(os.path.join(ibi_20_dir, f"Session{i+1:02d}", "Sess*.csv"))
  ibi_20_csv_files = sorted(ibi_20_csv_files)
  ibi_20_csv_array.append(ibi_20_csv_files)
  # eda
  eda_20_csv_files = glob.glob(os.path.join(eda_20_dir, f"Session{i+1:02d}", "Sess*.csv"))
  eda_20_csv_files = sorted(eda_20_csv_files)
  eda_20_csv_array.append(eda_20_csv_files)
  # temp
  temp_20_csv_files = glob.glob(os.path.join(temp_20_dir, f"Session{i+1:02d}", "Sess*.csv"))
  temp_20_csv_files = sorted(temp_20_csv_files)
  temp_20_csv_array.append(temp_20_csv_files)


# 파일 합치기
for i in range(40):
  # IBI
  ibi_20_df = pd.DataFrame(columns=['Speaker', 'IBI_mean', 'IBI_count', 'IBI_std', 'IBI_max', 'IBI_min', 'IBI_range', 'IBI_difference'])
  for file_session in ibi_20_csv_array[i]:
    ibi_20_csv_lines = []
    try:
      with open(file_session, 'r') as file:
        for line in file:
          values = line.strip().split(',')
          ibi_20_csv_lines.append(values)
      ibi_20 = pd.DataFrame(ibi_20_csv_lines)
      ibi_20 = ibi_20.iloc[1:, :]
      ibi_20.columns = ['Time_cum', 'IBI', 'Time', 'Speaker']
      ibi_20_grouped = signal_processing(ibi_20, 'IBI')
      ibi_20_df = pd.concat([ibi_20_df, ibi_20_grouped], ignore_index=True)
    except ValueError as e:
      pass
  # EDA
  eda_20_df = pd.DataFrame(columns=['Speaker', 'EDA_mean', 'EDA_count', 'EDA_std', 'EDA_max', 'EDA_min', 'EDA_range', 'EDA_difference'])
  for file_session in eda_20_csv_array[i]:
    eda_20_csv_lines = []
    try:
      with open(file_session, 'r') as file:
        for line in file:
          values = line.strip().split(',')
          eda_20_csv_lines.append(values)
      eda_20 = pd.DataFrame(eda_20_csv_lines)
      eda_20 = eda_20.iloc[2:, :]
      eda_20.columns = ['EDA', 'Time', 'Speaker']
      eda_grouped = signal_processing(eda_20, 'EDA')
      eda_20_df = pd.concat([eda_20_df, eda_grouped], ignore_index=True)
    except ValueError as e:
      pass
  # TEMP
  temp_20_df = pd.DataFrame(columns=['Speaker', 'TEMP_mean', 'TEMP_count', 'TEMP_std', 'TEMP_max', 'TEMP_min', 'TEMP_range', 'TEMP_difference'])
  for file_session in temp_20_csv_array[i]:
    temp_20_csv_lines = []
    try:
      with open(file_session, 'r') as file:
        for line in file:
          values = line.strip().split(',')
          temp_20_csv_lines.append(values)
      temp_20 = pd.DataFrame(temp_20_csv_lines)
      temp_20 = temp_20.iloc[2:, :]
      temp_20.columns = ['TEMP', 'Time', 'Speaker']
      temp_grouped = signal_processing(temp_20, 'TEMP')
      temp_20_df = pd.concat([temp_20_df, temp_grouped], ignore_index=True)
    except ValueError as e:
      pass
  # Annotation
  annotation_20 = pd.read_csv(annotation_20_files[i])
  annotation_20.rename(columns=new_columns_20, inplace=True)
  annotation_20 = annotation_20.iloc[1:, 1:]
  for e in evaluator_20:
    annotation_20[f'{e}_emotion'] = annotation_20[f'{e}_emotion'].map(emotion_category_map)
  for idx, row in annotation_20.iterrows():
    # Get P_i value of emotion evaluations
    emo_score = list(annotation_20.loc[idx, emotion_col_20])
    emo_tuple = [(0, e) for e in emo_score]
    annotation_20.loc[idx, 'Emotion_p_i'] = get_p_i(emo_tuple, 7, 10)

    # Get Fleiss-Kappa score of Valence & Arousal
    valence_score = list(map(int, list(annotation_20.loc[idx, valence_col_20])))
    valence_tuple = [(0, e) for e in valence_score]
    arousal_score = list(map(int, list(annotation_20.loc[idx, arousal_col_20])))
    arousal_tuple = [(1, e) for e in arousal_score]
    annotation_20.loc[idx, 'Fleiss_Kappa'] = get_kappa_score(valence_tuple + arousal_tuple, 2, 5, 10)
  annotation_20 = annotation_20[['Wav_start', 'Wav_end', 'Speaker', 'Emotion', 'Valence', 'Arousal', 'Emotion_p_i', 'Fleiss_Kappa']]

  # File merge
  merging_ibi_20_df = pd.merge(annotation_20, ibi_20_df, on='Speaker', how='left')
  merging_eda_20_df = pd.merge(merging_ibi_20_df, eda_20_df, on='Speaker', how='left')
  merged_df = pd.merge(merging_eda_20_df, temp_20_df, on='Speaker', how='left')
  
  # Add Session column
  current_file_name = annotation_20_files[i].split('/')[-1]
  current_file_name = current_file_name.split('.')[0]
  merged_df['Session'] = current_file_name

  # Concat to final dataframe
  final_20_df = pd.concat([final_20_df, merged_df], ignore_index=True)

  print(f'{current_file_name} ... Done!')
```
